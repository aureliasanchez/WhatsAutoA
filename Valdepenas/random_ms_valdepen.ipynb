```python
import pandas as pd
import random
```

### Lista de números y mensajes

```python
# Datos proporcionados
numeros = [
     "7711328731", "7711546065", "7711728731", "7713634153", "7721175511", "7751009373", "7751013130", "7751014226", "7751019464", "7751055775", "7751092638", "7751102231", "7751103295", "7751106326", "7751116940", "7751171636", "7751175511", "7751177719", "7751240630", "7751252975", "7751254749", "7751264628", "7751267172", "7751271720", "7751276583", "7751282871", "7751288817", "7751360568", "7751368081", "7751373790"     
    ]
# Mensajes con rutas de imágenes
mensajes = [
    {
        "mensaje": """📢 Le presento los *★Lotes Residenciales, Privada Tilo★*\n\n🎯Información general:\n- 🏡 Superficie: 105 m²\n- 📏Frente: 7m.\n- 📏Fondo: 15m.\n- 📍Tulancingo de Bravo, Hidalgo\n\n⛳Amenidades:\n- 🌲Área Verde\n- 🛝Juegos Infantiles\n- 🐾Zona de mascotas\n- 🏃🏻Pista de jogging\n- 🔐Seguridad\n\n💰 $525,000 MXN\n\n💡¿Te interesa obtener más información de este lote?""",
        "imagen": r"C:\Users\4to Creativo\OneDrive\Desktop\WhatsAutoA\Images\tilo.jpg"
    }
]
```

### Asignación random

```python
# Asignar un mensaje aleatorio a cada número
asignaciones = []
random.seed(42)  # Para reproducibilidad

for numero in numeros:
    asignacion = random.choice(mensajes)
    asignaciones.append({
        "Movil": numero,
        "Mensaje": asignacion["mensaje"].replace("\n", "\n"),
        "Imagen": asignacion["imagen"]
    })

# Convertir a DataFrame de pandas
df = pd.DataFrame(asignaciones)
df
```

### Exportar

```python
# Exportar a un archivo Excel
df.to_excel("asignaciones.xlsx", index=False)
```
